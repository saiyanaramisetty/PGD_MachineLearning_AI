```python
# Function to calculate and print all the evaluation metrics
def printEvaluationMetrics(y, y_pred, y_pred_proba, model, data,):
    cm = confusion_matrix(y, y_pred)
    print('Confusion Matrix: ', cm)
    tn, fp, fn, tp = confusion_matrix(y, y_pred).ravel()

    accuracy = round(metrics.accuracy_score(y_pred = y_pred, y_true = y) * 100, 2)
    print(f'Accuracy: {accuracy}') 

    precision = round(tp / (tp + fp), 2)
    print(f'Precision: {precision}') 

    recall = round(tp / (tp + fn) ,2)
    print(f'Recall: {recall}') 

    f1 = round((2 * (recall * precision) / (recall + precision)), 2)
    print(f'F1 Score: {f1}')
   
    fpr, tpr, thresholds = metrics.roc_curve(y, y_pred_proba)
    threshold = round(thresholds[np.argmax(tpr - fpr)], 4)
    print(f"Threshold: {threshold}")

    roc_auc = round(metrics.roc_auc_score(y, y_pred), 2)
    print(f'ROC AUC Score: {roc_auc}') 
    plt.plot(fpr, tpr, label = 'Score=' + str(roc_auc))
    plt.legend(loc = 4)
    plt.show()

    print(f'Classification Report: \n {classification_report(y, y_pred)}')

    print(pd.DataFrame({'Data': data, 'Accuracy': accuracy, 'ROC AUC Score': roc_auc, 'F1 Score': f1, 'Recall': recall, 'Precision': precision , 'Threshold': threshold, }, index = [model]))
```

```python
# Method for Logistic Regression model
def runLogisticRegression(X_train, X_test, y_train, y_test, model):
    folds = KFold(n_splits = 5, shuffle = True, random_state = 42)
    params = {'C': [0.01, 0.1, 1, 10, 100, 1000]}

    model_lr_cv = GridSearchCV(estimator = LogisticRegression(), param_grid = params, scoring= 'roc_auc', cv = folds, verbose = 2, return_train_score = True)
    model_lr_cv.fit(X_train, y_train)

    cv_results = pd.DataFrame(model_lr_cv.cv_results_)
    print(cv_results)
    print('Best Score:', model_lr_cv.best_score_)
    print('Best Hyperparameters:', model_lr_cv.best_params_)

    # Running with best hyperparameters
    model_lr_imbalance = LogisticRegression(C = model_lr_cv.best_params_['C'])
    model_lr_imbalance.fit(X_train, y_train)

    # Prediction on Train dataset
    print('Prediction on Train Dataset')
    y_train_pred = model_lr_imbalance.predict(X_train)
    y_train_pred_proba = model_lr_imbalance.predict_proba(X_train)[:, 1]
    printEvaluationMetrics(y_train, y_train_pred, y_train_pred_proba, model, 'Train')

    # Prediction on Test dataset
    print('Prediction on Test Dataset')
    y_test_pred = model_lr_imbalance.predict(X_test)
    y_test_pred_proba = model_lr_imbalance.predict_proba(X_test)[:,1]
    printEvaluationMetrics(y_test, y_test_pred, y_test_pred_proba, model, 'Test')

    return model_lr_imbalance
```

```python
# Method for XGBoost model
def runXGBoost(X_train, X_test, y_train, y_test, model):
    folds = 3
    param_grid = {'learning_rate': [0.2, 0.6], 'subsample': [0.3, 0.6, 0.9]}

    xgb_model = XGBClassifier(max_depth = 2, n_estimators = 200)
    model_xg_cv = GridSearchCV(estimator = xgb_model, param_grid = param_grid, scoring= 'roc_auc', cv = folds, verbose = 2, return_train_score = True)
    model_xg_cv.fit(X_train, y_train)
    
    cv_results = pd.DataFrame(model_xg_cv.cv_results_)
    print(cv_results)
    print('Best Score:', model_xg_cv.best_score_)
    print('Best Hyperparameters:', model_xg_cv.best_params_)
    print('Best Estimator:', model_xg_cv.best_estimator_)

    # Running with best hyperparameters
    params = {'learning_rate': model_xg_cv.best_params_['learning_rate'],'max_depth': 2, 'n_estimators' : 200,'subsample': model_xg_cv.best_params_['subsample'], 'objective':'binary:logistic'}
    model_xg_imbalance = XGBClassifier(params = params)
    model_xg_imbalance.fit(X_train, y_train)

    # Prediction on Train dataset
    print('Prediction on Train Dataset')
    y_train_pred = model_xg_imbalance.predict(X_train)
    y_train_pred_proba = model_xg_imbalance.predict_proba(X_train)[:, 1]
    printEvaluationMetrics(y_train, y_train_pred, y_train_pred_proba, model, 'Train')

    # Prediction on Test dataset
    print('Prediction on Test Dataset')
    y_test_pred = model_xg_imbalance.predict(X_test)
    y_test_pred_proba = model_xg_imbalance.predict_proba(X_test)[:, 1]
    printEvaluationMetrics(y_test, y_test_pred, y_test_pred_proba, model, 'Test')

    return model_xg_imbalance
```

```python
# Method for Decision Trees
def runDecisionTrees(X_train, X_test, y_train, y_test, model, criteria):
    folds = 3
    dtree = DecisionTreeClassifier()
    param_grid = {'max_depth': range(5, 15, 5),'min_samples_leaf': range(50, 150, 50),'min_samples_split': range(50, 150, 50)}
    model_dt_cv = GridSearchCV(estimator = dtree, param_grid = param_grid, scoring = 'roc_auc', cv = folds, verbose = 2)
    model_dt_cv.fit(X_train, y_train)
    
    cv_results = pd.DataFrame(model_dt_cv.cv_results_)
    print(cv_results)
    print('Best Score:', model_dt_cv.best_score_)
    print('Best Hyperparameters:', model_dt_cv.best_params_)
    print('Best Estimator:', model_dt_cv.best_estimator_)

    # Running with best hyperparameters
    model_dt_imbalance = DecisionTreeClassifier(criterion = criteria, max_depth = model_dt_cv.best_params_['max_depth'], min_samples_leaf = model_dt_cv.best_params_['min_samples_leaf'], min_samples_split = model_dt_cv.best_params_['min_samples_split'],  random_state = 42)
    model_dt_imbalance.fit(X_train, y_train)

    # Prediction on Train dataset
    print('Prediction on Train Dataset')
    y_train_pred = model_dt_imbalance.predict(X_train)
    y_train_pred_proba = model_dt_imbalance.predict_proba(X_train)[:, 1]
    printEvaluationMetrics(y_train, y_train_pred, y_train_pred_proba, model, 'Train')

    # Prediction on Test dataset
    print('Prediction on Test Dataset')
    y_test_pred = model_dt_imbalance.predict(X_test)
    y_test_pred_proba = model_dt_imbalance.predict_proba(X_test)[:,1]
    printEvaluationMetrics(y_test, y_test_pred, y_test_pred_proba, model, 'Test')
    
    return model_dt_imbalance
```
